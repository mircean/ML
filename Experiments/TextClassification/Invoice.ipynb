```python
file = r'C:\Users\mirce\OneDrive\Datasets\Classification\Receipts\dataset.csv'
df = pd.read_csv(file)
#df = df[['text', 'label', 'x1', 'y1', 'x2', 'y2']]
df = df[['text', 'label']]
df = df[df.label != 'O']
df = df[df.label != 'I']
np.random.seed(seed)
df_train, df_test = train_test_split(df, test_size=0.25, stratify=df.label)
data = dataset.Dataset(None, df_train, df_test)
data.df_train
```
