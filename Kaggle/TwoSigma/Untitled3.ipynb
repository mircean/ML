```python
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df1 = pd.read_csv('Output/Ensemble3/results_ensemble2.csv')
df1
```

```python
df2 = pd.read_csv('Output/Ensemble3/submission_053942.csv')
df2
```

```python
df_merge = df1.merge(df2, left_on='listing_id', right_on='listing_id')
df_merge
```

```python
percentage_1 = 50
percentage_2 = 100-percentage_1
df_merge['high'] = df_merge.high_x*percentage_1/100 + df_merge.high_y*percentage_2/100
df_merge['medium'] = df_merge.medium_x*percentage_1/100 + df_merge.medium_y*percentage_2/100
df_merge['low'] = df_merge.low_x*percentage_1/100 + df_merge.low_y*percentage_2/100
df_merge[['listing_id', 'high', 'medium', 'low']]
```

```python
df_merge[['high', 'medium', 'low', 'listing_id']].to_csv('Output/Ensemble3/results.csv', index=False)
```
